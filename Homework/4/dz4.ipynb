###  Домашнее задание к лекции "Функции"


```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def srch_num_name(): #Поиск ФИО по номеру
    number = input('Введите номер документа: ')
    for dic in documents:
        if number == dic['number']:
            return f"Владелец документа: {dic['name']}"
        else: pass
    return'Документ не найден в базе'


def srch_num_dir(number=None): #Поиск полки по номеру
    for element in directories.items():
        if number in element[1]:
            return f"Документ хранится на полке: {element[0]}"
        else: pass
    return f'Документ не найден в базе'
        

def full_info(): #Получение полной информации
    for element in documents:
        print(f"№: {element['number']}, тип: {element['type']}, владелец: {element['name']},\
 полка хранения: {srch_num_dir(element['number'])[-1]} ")
      
            
def ad_dir(): #Добавление полки
    number = input('Введите номер полки: ')
    if number in list(directories.keys()):
            return f'Такая полка уже существует. Текущий перечень полок: {", ".join(list(directories.keys()))}.'
    else: directories.update({number:[]})
    return f'Полка добавлена. Текущий перечень полок: {", ".join(list(directories.keys()))}.'
    

def dlt_dir(): #Удаление полки
    number = input('Введите номер полки: ')
    if number in list(directories.keys()):
        for element in directories.items():
            if element[0]==number and len(element[1])==0:
                del directories[number]
                return(f'Полка удалена. Текущий перечень полок: {", ".join(list(directories.keys()))}.') 
            else: pass
        return f'На полке есть документы, удалите их перед удалением полки.\
 Текущий перечень полок: {", ".join(list(directories.keys()))}.'
    else: return f'Такой полки не существует. Текущий перечень полок: {", ".join(list(directories.keys()))}.'   

def main():
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            print(srch_num_name())
        elif user_input == 's':
            number = input('Введите номер документа: ')
            print (srch_num_dir(number))
        elif user_input == 'l':
            full_info()
        elif user_input == 'ads':
            print(ad_dir())
        elif user_input == 'ds':
            print(dlt_dir())
        elif user_input == 'q':
            print('Завершение программы')
            break
        else: print('Неизвестная команда')

main()
```
